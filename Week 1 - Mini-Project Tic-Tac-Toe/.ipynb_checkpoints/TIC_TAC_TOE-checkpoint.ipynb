```python
import random

#Global
count_round=0
who_play='player'
round_max=9
victory=False #or True
tic_tac_toe=[
    [" "," "," "],
    [" "," "," "],
    [" "," "," "],
]

class color:
   BLUE = '\033[94m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   END = '\033[0m'
    
#Backgroud of game
def graph():
    """This function creates the game interface"""
    global tic_tac_toe
    global count_round
    print(color.BOLD +"ROUND "+ str(count_round)+ '\n' +color.END)    
    print("    0   1   2") #cabe√ßalho
    for i in range(0,3):
        print(str(i)+ ":  " + tic_tac_toe[i][0] + " | "+ tic_tac_toe[i][1] + " | " + tic_tac_toe[i][2]) 
        if i<2:
            print("   ------------")

#Player Turn
def PlayerTurn():
    """This function asks the user to write one coordinate (line and column).
    If empty, the field is filled with X. If not, one message appears and the function returns to beggining.
    
    Local variables: 
    l -line
    c- column"""
    global tic_tac_toe
    global count_round
    global who_play
    global round_max
    if ((who_play=='player') & (count_round<round_max) & (victory==False)):    

        try:
            print("Your Turn")
            l=int(input("Line: "))  
            c=int(input("Column:"))
            while((tic_tac_toe[l][c]=="X")|(tic_tac_toe[l][c]=="O")):
                print(color.RED+'Not empty!'+color.END)
                l=int(input("Line: "))  
                c=int(input("Column: "))
            tic_tac_toe[l][c]="X" 
            who_play='CPU'
            count_round=count_round+1            
  
    
        except:
            print(color.RED+"Not valid!"+color.END)
            PlayerTurn()

#CPU Turn
def CPUTurn():
    """This function choose one field to CPU round, using the randomn function.
    If empty, the field is filled with O. If not, other number is gerated.
 
    Local variables: 
    l -line
    c- column"""
    global tic_tac_toe
    global count_round
    global who_play
    global round_max
    global victory
    if ((who_play=='CPU') & (count_round<round_max) & (victory==False)):         
        print("CPU Turn")
        l=random.randrange(0,3)
        c=random.randrange(0,3)
        while((tic_tac_toe[l][c]!=" ")):
            l=random.randrange(0,3)
            c=random.randrange(0,3)
        tic_tac_toe[l][c]="O" 
        who_play='player'
        count_round=count_round+1   
        
#Verifying the lines
def WinLin():
    """This function checks if someone win, verifying if one of lines have 3 'X' or 3 'O'.
    
    Local variables: 
    l - line in range(0,3)
    
    Input: None
    
    Output: victory (True or False)"""
    
    global tic_tac_toe
    global victory
    for l in range(0,3):
        if ((tic_tac_toe[l][0]=="X") & (tic_tac_toe[l][1]=="X") & (tic_tac_toe[l][2]=="X")):
            print(color.BOLD+color.BLUE+'PLAYER WINS!'+color.END)
            victory=True
        if ((tic_tac_toe[l][0]=="O") & (tic_tac_toe[l][1]=="O") & (tic_tac_toe[l][2]=="O")):
            graph()
            print(color.BOLD+color.BLUE+'\nCPU WINS!'+color.END)  
            victory=True
    return victory

#Verifying the columns
def WinCol():
    """This function checks if someone win, verifying if one of columns have 3 'X' or 3 'O'.
    
    Local variables: 
    c- column in range(0,3)
    
    Input: None
    
    Output: victory (True or False)"""
    global tic_tac_toe
    global victory
    for c in range(0,3):
        if ((tic_tac_toe[0][c]=="X") & (tic_tac_toe[1][c]=="X") & (tic_tac_toe[2][c]=="X")):
            print(color.BOLD+color.BLUE+'PLAYER WINS!'+color.END)
            victory=True
        if ((tic_tac_toe[0][c]=="O") & (tic_tac_toe[1][c]=="O") & (tic_tac_toe[2][c]=="O")):
            graph()
            print(color.BOLD+color.BLUE+'\nCPU WINS!'+color.END)  
            victory=True
    return victory

#Verifying the diagonals
def WinDiag():
    """This function checks if someone win, verifying if one of diagonals have 3 'X' or 3 'O'.
    
    Input: None
    
    Output: victory (True or False)"""
    global tic_tac_toe
    global victory
    if ((tic_tac_toe[0][0]=="X") & (tic_tac_toe[1][1]=="X") & (tic_tac_toe[2][2]=="X")):
        print(color.BOLD+color.BLUE+'PLAYER WINS!'+color.END)
        victory=True
    if ((tic_tac_toe[0][0]=="O") & (tic_tac_toe[1][1]=="O") & (tic_tac_toe[2][2]=="O")):
        graph()
        print(color.BOLD+color.BLUE+'\nCPU WINS!'+color.END)  
        victory=True
    if ((tic_tac_toe[0][2]=="X") & (tic_tac_toe[1][1]=="X") & (tic_tac_toe[2][0]=="X")):
        print(color.BOLD+color.BLUE+'PLAYER WINS!'+color.END)
        victory=True
    if ((tic_tac_toe[0][2]=="O") & (tic_tac_toe[1][1]=="O") & (tic_tac_toe[2][0]=="O")):
        graph()
        print(color.BOLD+color.BLUE+'\nCPU WINS!'+color.END)  
        victory=True
    return victory

# Main program
while True: 
    graph()
    print('\n')
    PlayerTurn()
    graph()
    print('\n')
    victory=WinLin()
    victory=WinCol()
    victory=WinDiag()
    if victory==False:
        CPUTurn()
        victory=WinLin()
        victory=WinCol()
        victory=WinDiag()
    if ((victory== False)&(count_round==round_max)):
        print(color.BOLD+color.BLUE+"TIC-TAC-TOE!"+color.END)
    if (victory==True)|(count_round>=round_max):
        break
```
