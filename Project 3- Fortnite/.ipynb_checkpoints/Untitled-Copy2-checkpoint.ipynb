# LEADERBOARD FORTNITE WORLD CUP FINALS

```python
import os
import requests
import pandas as pd
import numpy as np
from dotenv import load_dotenv, find_dotenv
from bs4 import BeautifulSoup
```

## Web scrap to get top 25

```python
response= requests.get('https://en.wikipedia.org/wiki/Fortnite_World_Cup')
html = response.content
soup = BeautifulSoup(html, "lxml")
table_25 = pd.read_html(html)
table_25 = table_25[1]
table_25
```

```python
#Data Cleaning
table_25['Place'] = [i.replace('st','')for i in table_25['Place']]
table_25['Place'] = [i.replace('nd','')for i in table_25['Place']]
table_25['Place'] = [i.replace('rd','')for i in table_25['Place']]
table_25['Place'] = [i.replace('th','')for i in table_25['Place']]

#Removing $ symbol
table_25['Prize Money'] = [i.replace('$','')for i in table_25['Prize Money']]

#Renaming column Prize Money
table_25.rename(columns={'Prize Money':'Reward Dollars'}, inplace = True)

#Renaming column Team 
table_25.rename(columns={'Team':'Player ID'}, inplace = True)

#Modifying KING and Psalm name (to match with API result)
table_25.loc[4,'Player ID']='King'
table_25.loc[1,'Player ID']='Psalm'

#Check result
table_25
```

```python
#Applying the API for each Player ID
url = [f'https://fortnite-api.com/v1/stats/br/v2?name={name}' for name in table_25['Player ID']]
response = [requests.get(i).json() for i in url]
```

```python
table_25_api = pd.json_normalize(response)
table_25_api 
pd.set_option('display.max_columns', None)
```

```python
#Drop Columns that won't be used
columns_to_drop  = list(table_25_api.loc[:,[i.startswith('stats') for i in list(table_25_api.columns)]].columns)
table_25_api.drop(columns = columns_to_drop, inplace= True)
table_25_api.drop(columns = ['trio','squad','ltm','solo','duo'],inplace= True)
```

```python
table_25_api
```

```python
#MAIN TABLE - Merge the Web scrap with API content
top_25_info = pd.merge(left=table_25, right=table_25_api, left_on='Player ID', right_on='account name')
```

```python
top_25_info.rename(columns={'Place':'Rank Position'}, inplace=True)
top_25_info.drop(columns=['error','status','account name'], inplace=True)
```

```python
top_25_info
```

```python
#Top players - Export to CSV
top_25_info.to_csv('./Dataset/top_25_info.csv', index=False)
```

## Check the quantity of private accounts

```python
#Merge the Web scrap with API content - LEFT JOIN
top_25_info2 = pd.merge(left=table_25, right=table_25_api, left_on='Player ID', right_on='account name', how='left')
```

```python
top_25_info2.rename(columns={'Place':'Rank Position'}, inplace=True)
```

```python
top_25_info2.loc[top_25_info2['status'].isnull() == False, 'status'] = 'Available'
top_25_info2.loc[top_25_info2['status'].isnull() == True , 'status'] = 'Private'
top_25_info2.loc[((top_25_info2['Player ID']!='Psalm') | (top_25_info2['Player ID']!='King') | (top_25_info2['Player ID']!='BELAEU') | (top_25_info2['Player ID']!='Commandment') | (top_25_info2['Player ID']!='Bizzle')) & (top_25_info2['status']!='Private'), 'status'] = 'Not Allowed'
```

```python
top_25_info2.drop(columns=['error','account name'], inplace=True)
```

```python
top_25_info2
```

```python
#Top players With Status - Export to CSV
top_25_info2.to_csv('./DataSet/top_25_status_available.csv', index=False)
```

## MY STATS

```python
name = ['Floelr']
results = [requests.get(f'https://fortnite-api.com/v1/stats/br/v2?name={i}').json() for i in name]
me = pd.json_normalize(results)
columns_to_drop = list(me.loc[: ,me.isnull().mean()>0.8].columns)
me.drop(columns=columns_to_drop, inplace=True)
me.drop(columns='status', inplace=True)

rename_columns = [(' '.join(i.split(sep='.'))).replace('data stats all','') for i in me.columns]
rename_columns2 = [i.replace('data','') for i in rename_columns]

#Removing extra spaces:
rename_columns3 = [i.lstrip() for i in rename_columns2]

me.columns= rename_columns3

me
```

```python
#Top players With Status - Export to CSV
me.to_csv('./DataSet/my_informations.csv', index=False)
```
