## Web scrap to get top 25

```python
response= requests.get('https://en.wikipedia.org/wiki/Fortnite_World_Cup')
html = response.content
soup = BeautifulSoup(html, "lxml")
table_25 = pd.read_html(html)
table_25 = table_25[1]
table_25
```

```python
#Data Cleaning
table_25['Place'] = [i.replace('st','')for i in table_25['Place']]
table_25['Place'] = [i.replace('nd','')for i in table_25['Place']]
table_25['Place'] = [i.replace('rd','')for i in table_25['Place']]
table_25['Place'] = [i.replace('th','')for i in table_25['Place']]

#Removing $ symbol
table_25['Prize Money'] = [i.replace('$','')for i in table_25['Prize Money']]

#Renaming column Prize Money
table_25.rename(columns={'Prize Money':'Reward Dollars'}, inplace = True)

#Renaming column Team 
table_25.rename(columns={'Team':'Player ID'}, inplace = True)

#Check result
table_25
```

```python
# Get only TOP 10 
table_10 = table_25.drop(table_25.index[range(11,25)])
table_10
```
