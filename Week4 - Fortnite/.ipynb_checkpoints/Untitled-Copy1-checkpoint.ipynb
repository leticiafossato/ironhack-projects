```python
import os
import requests
import pandas as pd
import numpy as np
from dotenv import load_dotenv, find_dotenv
from bs4 import BeautifulSoup
```

```python
#https://fortnitetracker.com/events/epicgames_ONSITE_WorldCup_Solos
```

## Web scrap to get top 25

```python
response= requests.get('https://en.wikipedia.org/wiki/Fortnite_World_Cup')
html = response.content
soup = BeautifulSoup(html, "lxml")
table_25 = pd.read_html(html)
table_25 = table_25[1]
table_25
```

```python
#Data Cleaning
table_25['Place'] = [i.replace('st','')for i in table_25['Place']]
table_25['Place'] = [i.replace('nd','')for i in table_25['Place']]
table_25['Place'] = [i.replace('rd','')for i in table_25['Place']]
table_25['Place'] = [i.replace('th','')for i in table_25['Place']]

#Removing $ symbol
table_25['Prize Money'] = [i.replace('$','')for i in table_25['Prize Money']]

#Renaming column Prize Money
table_25.rename(columns={'Prize Money':'Reward Dollars'}, inplace = True)

#Renaming column Team 
table_25.rename(columns={'Team':'Player ID'}, inplace = True)

#Modifying KING and Psalm name (to match with API result)
table_25.loc[4,'Player ID']='King'
table_25.loc[1,'Player ID']='Psalm'

#Check result
table_25
```

```python
# Get only TOP 10 
table_10 = table_25.drop(table_25.index[range(11,25)])
table_10
```

```python
table_10
```

```python
#Applying the API for each Player ID
url = [f'https://fortnite-api.com/v1/stats/br/v2?name={name}' for name in table_10['Player ID']]
response = [requests.get(i).json() for i in url]
```

```python
table_10_api = pd.json_normalize(response)
table_10_api 
pd.set_option('display.max_columns', None)
```

```python
table_10_api
```

```python
#Create a backup
table_10_api_bkp = table_10_api.copy()
```

```python
#Rename columns
rename_columns = [(' '.join(i.split(sep='.'))).replace('data stats all','') for i in table_10_api.columns]
rename_columns2 = [i.replace('data','') for i in rename_columns]
rename_columns3 = [i.lstrip() for i in rename_columns2]
table_10_api.columns= rename_columns3
```

```python
table_10_api
```

```python
#Drop Columns that won't be used
columns_to_drop  = list(table_10_api.loc[:,[i.startswith('stats') for i in list(table_10_api.columns)]].columns)
table_10_api.drop(columns = columns_to_drop, inplace= True)
table_10_api.drop(columns = ['trio','squad','ltm','solo','duo'],inplace= True)
```

```python
table_10_api
```

```python
table_10_api.columns
```

```python
#MAIN TABLE - Merge the Web scrap with API content
top_10_info = pd.merge(left=table_10, right=table_10_api, left_on='Player ID', right_on='account name')
```

```python
top_10_info.rename(columns={'Place':'Rank Position'}, inplace=True)
top_10_info.drop(columns=['error','status','account name'], inplace=True)
```

```python
top_10_info
```

```python
#Top players - Export to CSV
top_10_info.to_csv('./Datasets/top_10_info.csv', index=False)
```

## Check the quantity of private accounts

```python
#Merge the Web scrap with API content - LEFT JOIN
top_10_info2 = pd.merge(left=table_10, right=table_10_api, left_on='Player ID', right_on='account name', how='left')
```

```python
top_10_info2.rename(columns={'Place':'Rank Position'}, inplace=True)
top_10_info2
```

```python
top_10_info2['status'].loc[top_10_info2['status'].isnull() == False] = 'Available'
top_10_info2['status'].loc[top_10_info2['status'].isnull() == True] = 'Private'
```

```python
top_10_info2.drop(columns=['error','account name'], inplace=True)
```

```python
top_10_info2
```

```python
#Top players With Status - Export to CSV
top_10_info2.to_csv('./Datasets/top_10_status_available.csv', index=False)
```
